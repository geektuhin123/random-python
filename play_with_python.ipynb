```python
demo = 'hello world! 123'
def count_digits_string(input):
    digit_count = 0
    char_count = 0
    for i in demo:
        if(i.isdigit()):
            digit_count += digit_count
        if(i.isalpha()):
            char_count += char_count
    return ("digit": digit_count, "char": char_count)

count_digits_string(demo)
```
